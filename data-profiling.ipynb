```python
import pandas as pd
import numpy as np
import cv2
import ast
import os

df = pd.read_csv('.\\data\\H_trajectories\\phone_camera-image_color-compressed.csv')

output_dir = 'obrazy_z_csv_phone'
os.makedirs(output_dir, exist_ok=True)

for idx, row in df.iterrows():
    try:
        byte_data = ast.literal_eval(row['data'])
        np_arr = np.frombuffer(byte_data, dtype=np.uint8)
        image = cv2.imdecode(np_arr, cv2.IMREAD_COLOR)

        if image is not None:
            filename = os.path.join(output_dir, f'obraz_{idx:05d}.jpg')
            cv2.imwrite(filename, image)
            print(f'Zapisano: {filename}')
        else:
            print(f'Błąd dekodowania obrazu w wierszu {idx}')

    except Exception as e:
        print(f'Błąd w wierszu {idx}: {e}')
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(".\\data\\H_trajectories\\orb_slam3_phone.csv")

print(df.info())

corr_matrix = df.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=False, cmap='Reds', fmt=".2f", square=True, cbar=True)
plt.title("Macierz Korelacji orb_slam3_phone")
plt.tight_layout()
plt.show()
```

TESTOWY PLIK:

```python
import bagpy
from bagpy import bagreader
import pandas as pd

bag = bagreader('.\\data\\G_trajectories.bag')

# dostępne topiki
topics = list(bag.topics)
print(topics)

for topic in topics:
    csv_path = bag.message_by_topic(topic)
    df = pd.read_csv(csv_path)
    print(topic)
    print(df.columns)
    print(df.head())
```
